```python
#Function based Programming
def check_ingredients(res, ingredients, kopi_name):
    for ingr in res.keys():
        if kopi_name in ingredients and \
            ingr in ingredients[kopi_name] and \
            res[ingr] < ingredients[kopi_name][ingr]:
            return False
    return True    

def calculate_coins(ten_cents=0, twenty_cents=0, fifty_cents=0, one_dollar=0):
    return ten_cents*10 + twenty_cents*20 + fifty_cents*50 + one_dollar*100

def print_resources(res):
    print('Available resources: ')
    for ingr, amnt in res.items():
        print(ingr+':', amnt)

def print_menu(menu, ingredients, res):
    print('Available selections: ')
    for i in range(len(menu)):
        if all(res[ingr] >= ingredients[menu[i]][ingr] for ingr in ingredients[menu[i]]):
            print('{0}: {1}'.format(i, menu[i]))

def print_order(order):
    print('Your ordered a {0}.'.format(order['item']))

def make_coffee(res, ingredients, order):
    if check_ingredients(res, ingredients, order['item']):
        print('Preparing your kopi ...')
        for ingr in ingredients[order['item']]:
            res[ingr] -= ingredients[order['item']][ingr]
        print('Your kopi is ready!')
        return True
        
def serve_orders(orders):
    resources = {
        'water': MAX_WATER,
        'coffee': MAX_COFFEE,
        'milk': MAX_MILK,
        'sugar': MAX_SUGAR
    }
    machine_on = True
    payment_collected = 0
    
    while machine_on:
        print()
        if orders:
            ord1 = orders.pop(0)
            if ord1['item'].upper() in INGREDIENTS:
                print_order(ord1)
                paid = calculate_coins(**ord1['coins'])
                print('You paid {0} dollars {1} cents.'.format(paid//100, paid%100))
                to_pay = PRICES[ord1['item']]
                if paid < to_pay:
                    print('Not enough payment. Please collect your coins.')
                    print('Order is skipped.')
                    continue
                elif paid >= to_pay:
                    if paid > to_pay:
                        change = paid - to_pay
                        print('Your change is {0} dollars {1} cents.'.format(change//100, change%100))
                    payment_collected += to_pay
                    if not check_ingredients(resources, INGREDIENTS, ord1['item']):
                        print('Not enough resources. Refilling ...')
                        resources['water'] = MAX_WATER
                        resources['coffee'] = MAX_COFFEE
                        resources['milk'] = MAX_MILK
                        resources['sugar'] = MAX_SUGAR
                    make_coffee(resources, INGREDIENTS, ord1)

            elif ord1['item'].lower() == 'off':
                print('Turning off machine ...')
                machine_on = False
            elif ord1['item'].lower() == 'show menu':
                print_menu(MENU, INGREDIENTS, resources)
            elif ord1['item'].lower() == 'show resources':
                print_resources(resources)
            elif ord1['item'].lower() == 'show total payment':
                print('{0} dollars {1} cents has been collected.'.format(payment_collected//100, 
                                                                         payment_collected%100))
            else:
                pass
        else:
            print('All orders are processed.')
            machine_on = False


MENU = ['KOPI-O', 'KOPI-C', 'KOPI-C SIEW DAI', 'KOPI-C KOSONG']
INGREDIENTS = {'KOPI-O': {'water': 50, 'coffee': 15},
               'KOPI-C': {'water': 40, 'coffee': 10, 'milk': 10, 'sugar': 10},
               'KOPI-C SIEW DAI': {'water': 40, 'coffee': 10, 'milk': 10, 'sugar': 5},
               'KOPI-C KOSONG': {'water': 40, 'coffee': 10, 'milk': 10}
              }
PRICES = { 'KOPI-O': 150,
           'KOPI-C': 180,
           'KOPI-C SIEW DAI': 170,
           'KOPI-C KOSONG': 160
         }
MAX_WATER = 200
MAX_COFFEE = 100
MAX_MILK = 50
MAX_SUGAR = 20
            
all_order = [
    {'item': 'KOPI-C',
     'coins': {'ten_cents': 1, 'twenty_cents': 1, 'fifty_cents': 3}},
    {'item': 'KOPI-O',
     'coins': {'ten_cents': 5, 'twenty_cents': 1, 'fifty_cents': 1}},
    {'item': 'show total payment'},
    {'item': 'KOPI-C',
     'coins': {'ten_cents': 1, 'twenty_cents': 4, 'fifty_cents': 2}},
    {'item': 'show resources'},
    {'item': 'KOPI-C SIEW DAI',
     'coins': {'ten_cents': 1, 'twenty_cents': 1, 'fifty_cents': 3}},
    {'item': 'show menu'},
    {'item': 'off'},
    {'item': 'show resources'},
    {'item': 'KOPI-C KOSONG',
     'coins': {'ten_cents': 1, 'twenty_cents': 4, 'fifty_cents': 2}}
]

serve_orders(all_order)

all_order2 = [
    {'item': 'KOPI-O',
     'coins': {'ten_cents': 1, 'twenty_cents': 1, 'fifty_cents': 3}},
    {'item': 'KOPI-C',
     'coins': {'one_dollar': 1, 'twenty_cents': 2, 'fifty_cents': 1}},
    {'item': 'KOPI-C KOSONG',
     'coins': {'ten_cents': 1, 'twenty_cents': 4, 'fifty_cents': 2}},
    {'item': 'show resources'},
    {'item': 'KOPI-C',
     'coins': {'ten_cents': 1, 'twenty_cents': 1, 'fifty_cents': 3}},
    {'item': 'show menu'},
    {'item': 'show resources'},
    {'item': 'show total payment'},
    
]

serve_orders(all_order2)

all_order3 = [
    {'item': 'KOPI-O',
     'coins': {'ten_cents': 1, 'twenty_cents': 1, 'fifty_cents': 3}},
    {'item': 'KOPI-C',
     'coins': {'one_dollar': 1, 'twenty_cents': 2, 'fifty_cents': 1}},
    {'item': 'KOPI-C KOSONG',
     'coins': {'ten_cents': 1, 'twenty_cents': 4, 'fifty_cents': 2}},
    {'item': 'show resources'},
    {'item': 'KOPI-C',
     'coins': {'ten_cents': 1, 'twenty_cents': 1, 'fifty_cents': 3}},
    {'item': 'show menu'},   
    {'item': 'KOPI-C SIEW DAI',
     'coins': {'twenty_cents': 5, 'one_dollar': 1}},
    {'item': 'show total payment'},
    {'item': 'KOPI-O',
     'coins': {'fifty_cents': 4}},
    {'item': 'show resources'}
]

serve_orders(all_order3)

all_order4 = [
    {'item': 'KOPI-C SIEW DAI',
     'coins': {'ten_cents': 1, 'twenty_cents': 1, 'fifty_cents': 3}}, 
    {'item': 'KOPI-O',
     'coins': {'one_dollar': 1, 'twenty_cents': 1, 'fifty_cents': 1}},
    {'item': 'KOPI-C KOSONG',
     'coins': {'ten_cents': 1, 'twenty_cents': 4, 'fifty_cents': 2}},
    {'item': 'show total payment'},
    {'item': 'KOPI-C KOSONG',
     'coins': {'ten_cents': 1, 'twenty_cents': 1, 'one_dollar': 1}},
    {'item': 'show menu'},
    {'item': 'KOPI-C SIEW DAI',
     'coins': {'twenty_cents': 5, 'one_dollar': 1}},
    {'item': 'show total payment'},
    {'item': 'KOPI-O',
     'coins': {'fifty_cents': 4}},
    {'item': 'show resources'},
    {'item': 'show total payment'},
    {'item': 'off'},
]

serve_orders(all_order4)
```

```python
# cell 2
# works to produce the same output in cell 3 as in the output in cell 1

class CoffeeMachine:
    def __init__(self, menu, prices, ingredients, max_resources):
        self.menu = menu
        self.prices = prices
        self.ingredients = ingredients
        self.max_resources = max_resources   
        
    def print_order(self, orders):
        print('Your ordered a {0}.'.format(orders['item']))   
        
    def calculate_coins(self, ten_cents=0, twenty_cents=0, fifty_cents=0, one_dollar=0):
        return ten_cents*10 + twenty_cents*20 + fifty_cents*50 + one_dollar*100  

    def check_ingredients(self, res, kopi_name):
        for ingr in res.keys():
            if kopi_name in self.ingredients and \
                ingr in self.ingredients[kopi_name] and \
                res[ingr] < self.ingredients[kopi_name][ingr]:
                return False
        return True    

    def make_coffee(self, res, order):
        if self.check_ingredients(res, order['item']):
            print('Preparing your kopi ...')
            for ingr in self.ingredients[order['item']]:
                res[ingr] -= self.ingredients[order['item']][ingr]
            print('Your kopi is ready!')
            return True

    def print_resources(self,res):
        print('Available resources: ')
        for ingr, amnt in res.items():
            print(ingr+':', amnt)

    def print_menu(self,  res):
        print('Available selections: ')
        for i in range(len(self.menu)):
            if all(res[ingr] >= self.ingredients[self.menu[i]][ingr] for ingr in self.ingredients[self.menu[i]]):
                print('{0}: {1}'.format(i, self.menu[i]))
            
    def serve_orders(self, orders):
        resources = {
            'water': self.max_resources['water'],
            'coffee': self.max_resources['coffee'],
            'milk': self.max_resources['milk'],
            'sugar': self.max_resources['sugar']
        }
        machine_on = True
        payment_collected = 0

        while machine_on:
            print()
            if orders:
                ord1 = orders.pop(0)
                if ord1['item'].upper() in self.ingredients:
                    self.print_order(ord1)
                    paid = self.calculate_coins(**ord1['coins'])
                    print('You paid {0} dollars {1} cents.'.format(paid//100, paid%100))
                    to_pay = PRICES[ord1['item']]
                    if paid < to_pay:
                        print('Not enough payment. Please collect your coins.')
                        print('Order is skipped.')
                        continue
                    elif paid >= to_pay:
                        if paid > to_pay:
                            change = paid - to_pay
                            print('Your change is {0} dollars {1} cents.'.format(change//100, change%100))
                        payment_collected += to_pay
                        if not self.check_ingredients(resources, ord1['item']):
                            print('Not enough resources. Refilling ...')
                            resources['water'] = self.max_resources['water']
                            resources['coffee'] = self.max_resources['coffee']
                            resources['milk'] = self.max_resources['milk']
                            resources['sugar'] = self.max_resources['sugar']
                        self.make_coffee(resources, ord1)

                elif ord1['item'].lower() == 'off':
                    print('Turning off machine ...')
                    machine_on = False
                elif ord1['item'].lower() == 'show menu':
                    self.print_menu(resources)
                elif ord1['item'].lower() == 'show resources':
                    self.print_resources(resources)
                elif ord1['item'].lower() == 'show total payment':
                    print('{0} dollars {1} cents has been collected.'.format(payment_collected//100, 
                                                                             payment_collected%100))
                else:
                    pass
            else:
                print('All orders are processed.')
                machine_on = False
```
